### 2.1. Criação da conexão com o banco
* No momento da conexão, é informado o nome do arquivo onde será montado o banco de dados. No nosso caso, este arquivo não precisa ser pré-existente, porque iremos carregar as informações de um dump;
* Se este comando for executado com um arquivo que já existe e possui as mesmas tabelas que serão criadas, ocorrerá um erro no momento da execução do script mais à frente.

````python
conn = sqlite3.connect('despesas.db')
cursor = conn.cursor()
````

### 2.2. Execução do script para inicializar e popular o banco de dados

* Criada a conexão, iremos utilizar o arquivo .sql disponibilizado para recuperar o schema e os dados do banco

````python
f = io.open('despesas_publicas_tp1.sql', 'r', encoding='ISO-8859-1')
sql = f.read()
cursor.executescript(sql)
````

**2.** Dado a coluna nome_orgao, retorne a primeira posição que a string 'da' aparece da tabela orgão 

````python
df = pd.read_sql_query("SELECT NOME_ORGAO, instr(NOME_ORGAO, 'da')  AS DA_POSICAO FROM ORGAO", conn)
df
````

**3.** Liste todos os nomes dos órgãos e o valor das despesas totais no ano de 2018 para os mesmos, ordenando de forma decrescente pelo valor.

````python
df = pd.read_sql_query("SELECT O.NOME_ORGAO, SUM(D.VALOR_DESPESA) as VALOR_DESPESA " + 
                       "FROM DESPESA_PUBLICA AS D, ORGAO AS O " +
                       "WHERE D.COD_ORGAO = O.COD_ORGAO AND D.ANO_DESPESA=2018 " + 
                       "GROUP BY O.NOME_ORGAO " +
                       "ORDER BY VALOR_DESPESA DESC", conn)
df

````

**4.** Liste o mês e ano, modalidade, programa de governo e valor de todas as despesas do órgão Ministério da Fazenda no primeiro semestre de 2018.

> Bloco com recuo



````python
df = pd.read_sql_query("SELECT MES_DESPESA, ANO_DESPESA, MODALIDADE_DESPESA, P.NOME_PROGRAMA_GOVERNO, SUM(VALOR_DESPESA) AS VALOR_TODAS_DESPESAS " + 
                       "FROM DESPESA_PUBLICA AS D "+
                       "INNER JOIN PROGRAMA_GOVERNO AS P "+
                       "ON D.COD_PROGRAMA_GOVERNO=P.COD_PROGRAMA_GOVERNO "+
                       "INNER JOIN ORGAO AS O " +
                       "ON O.COD_ORGAO = D.COD_ORGAO "+
                       "WHERE O.NOME_ORGAO='Ministério da Fazenda' AND MES_DESPESA<=6 AND D.ANO_DESPESA=2018 "+
                       "GROUP BY MODALIDADE_DESPESA, MES_DESPESA, ANO_DESPESA, P.NOME_PROGRAMA_GOVERNO", conn)
df
````

**5.** Liste o código e nome dos programas de governo que possuíram os 3 maiores valores de despesas totais na função de educação durante o ano de 2018.

````python
df = pd.read_sql_query("SELECT P.COD_PROGRAMA_GOVERNO, P.NOME_PROGRAMA_GOVERNO " + 
                       "FROM DESPESA_PUBLICA AS D " +
                       "INNER JOIN DETALHAMENTO_FUNCAO AS DF "+
                       "ON D.COD_DETALHAMENTO=DF.COD_DETALHAMENTO "+
                       "INNER JOIN FUNCAO AS F "+
                       "ON DF.COD_FUNCAO=F.COD_FUNCAO "+
                       "INNER JOIN PROGRAMA_GOVERNO AS P "+
                       "ON D.COD_PROGRAMA_GOVERNO=P.COD_PROGRAMA_GOVERNO "+ 
                       "WHERE F.NOME_FUNCAO='Educação' AND D.ANO_DESPESA=2018 "+
                       "GROUP BY P.COD_PROGRAMA_GOVERNO " +
                       "ORDER BY SUM(D.VALOR_DESPESA) DESC LIMIT 3 ", conn)
df
````

**6.** Transforme os caracteres em minúsculo e remova todos os termos 'da' na coluna nome_orgao da tabela orgão.

````python
df = pd.read_sql_query("SELECT REPLACE(LOWER(NOME_ORGAO), 'da','') AS NOME_ORGAO_SEM_DA FROM ORGAO", conn)
df
````

**7.** Liste os órgãos governamentais e a quantidade de programas de governo distintos envolvidos nas despesas públicas desses órgãos. Nesta consulta, desconsidere valores de despesa negativos ou iguais a zero no período.

````python
df = pd.read_sql_query("SELECT NOME_ORGAO, COUNT(DISTINCT P.COD_PROGRAMA_GOVERNO) AS QTD_PROGRAMAS_DISTINTOS " + 
                       "FROM ORGAO AS O, DESPESA_PUBLICA AS D "+
                       "INNER JOIN PROGRAMA_GOVERNO AS P "+
                       "ON D.COD_PROGRAMA_GOVERNO=P.COD_PROGRAMA_GOVERNO "+
                       "WHERE O.COD_ORGAO=D.COD_ORGAO AND D.VALOR_DESPESA>0 "+
                       "GROUP BY NOME_ORGAO", conn)
df
````

**8.** Liste os nomes das subfunções das despesas públicas envolvidas para cada programa de governo. Sua consulta deve retornar o nome do programa de governo, a subfunção da despesa, a quantidade de entradas de despesas públicas e o valor total dessas despesas.

````python
df = pd.read_sql_query("SELECT NOME_PROGRAMA_GOVERNO, NOME_SUBFUNCAO, COUNT(VALOR_DESPESA), SUM(VALOR_DESPESA)" + 
                       "FROM DESPESA_PUBLICA AS D "+
                       "INNER JOIN DETALHAMENTO_FUNCAO AS DF "+
                       "ON D.COD_DETALHAMENTO=DF.COD_DETALHAMENTO "+
                       "INNER JOIN SUBFUNCAO AS S "+
                       "ON DF.COD_SUBFUNCAO=S.COD_SUBFUNCAO "+
                       "INNER JOIN PROGRAMA_GOVERNO AS P "+
                       "ON D.COD_PROGRAMA_GOVERNO=P.COD_PROGRAMA_GOVERNO "+
                       "GROUP BY NOME_PROGRAMA_GOVERNO, NOME_SUBFUNCAO", conn)
df
````

**9.** Liste todas as despesas públicas (código da despesa, mês e ano da despesa, modalidade, valor) em que o órgão responsável é aquele que possui o maior número de órgãos subordinados.

````python
df = pd.read_sql_query("SELECT COD_DESPESA, MES_DESPESA, ANO_DESPESA, MODALIDADE_DESPESA, SUM(VALOR_DESPESA) " + 
                       "FROM DESPESA_PUBLICA AS D "+
                       "INNER JOIN ORGAO AS O "+
                       "ON D.COD_ORGAO=O.COD_ORGAO "+
                       "WHERE O.QTD_ORGAOS_SUBORDINADOS=(SELECT MAX(QTD_ORGAOS_SUBORDINADOS)FROM ORGAO) "+
                       "GROUP BY D.COD_DESPESA ", conn)
df
````

**10.** Liste todas as despesas públicas (código da despesa, mês e ano da despesa, modalidade, valor) em que o órgão responsável é aquele que possui o maior valor total em despesas durante o ano de 2018.

````python
df = pd.read_sql_query("SELECT COD_DESPESA, MES_DESPESA, ANO_DESPESA, MODALIDADE_DESPESA, VALOR_DESPESA, D.COD_ORGAO " + 
                       "FROM DESPESA_PUBLICA AS D "+
                       "INNER JOIN ORGAO AS O "+
                       "ON D.COD_ORGAO=(SELECT O.COD_ORGAO FROM ORGAO AS O "+
                                        "INNER JOIN DESPESA_PUBLICA AS D "+
                                        "ON O.COD_ORGAO=D.COD_ORGAO "+
                                        "WHERE D.ANO_DESPESA=2018 "+
                                        "GROUP BY O.COD_ORGAO "+
                                        "ORDER BY SUM(VALOR_DESPESA) DESC LIMIT 1) "+
                       "GROUP BY D.COD_DESPESA, MES_DESPESA, ANO_DESPESA ", conn)
df

````

**11.** Liste os códigos e nomes dos programas de governo relacionados ao órgão (ou órgãos, caso haja empate) que teve menos registros de despesas públicas durante o ano.

````python
df = pd.read_sql_query("SELECT D.COD_PROGRAMA_GOVERNO, NOME_PROGRAMA_GOVERNO FROM (SELECT COD_ORG FROM (SELECT COD_ORGAO AS COD_ORG, "+
                                                                                   "COUNT(D.VALOR_DESPESA) COUNT_DESPESAS, SUM(D.VALOR_DESPESA) "+
                                                                                   "SUM_DESPESA FROM DESPESA_PUBLICA AS D "+
                                                                                    "GROUP BY D.ANO_DESPESA, D.COD_ORGAO) "+
                       "WHERE COUNT_DESPESAS = (SELECT COUNT(D.VALOR_DESPESA) C_DESPESAS FROM DESPESA_PUBLICA AS D "+
                                                "GROUP BY D.ANO_DESPESA, D.COD_ORGAO ORDER BY C_DESPESAS ASC LIMIT 1) GROUP BY COD_ORG) "+
                       "INNER JOIN DESPESA_PUBLICA AS D "+
                       "ON COD_ORG=D.COD_ORGAO "+
                       "INNER JOIN PROGRAMA_GOVERNO AS P "+
                       "ON D.COD_PROGRAMA_GOVERNO=P.COD_PROGRAMA_GOVERNO", conn)
df

````
